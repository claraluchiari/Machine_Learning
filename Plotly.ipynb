```python
import plotly.graph_objects as go
```

```python
anos = [2018, 2019, 2020, 2021, 2022]
vendas = [100, 150, 200, 180, 220]
```

```python
from re import X

fig = go.Figure(data=go.Scatter(x = anos, y = vendas))
fig.update_layout(
    title = "Vendas pelo tempo",
    xaxis_title = "Ano",
    yaxis_title = "Vendas",
    showlegend = False
)

fig.add_annotation(
    x = 2020,
    y = 200,
    text = "Maiores vendas",
    showarrow = True,
    arrowhead = 1,
    ax = -50,
    ay = -50
)

fig.show()
```

```python
idades = [18, 20, 22, 19, 25, 21]
scores = [78, 92, 88, 75, 86, 90]
```

```python
# Scatter Plot

fig = go.Figure(data = go.Scatter(x = idades, y = scores, mode = "markers"))
fig.update_layout(
    title = "Idades x scores",
    xaxis_title = "Idades",
    yaxis_title = "Scores",
    showlegend = False
)
fig.show()
```

```python
fig = go.Figure(data = go.Scatter(x = idades, y = scores, mode="markers", marker = dict(
    symbol = "star",
    size = 10,
    color = "red",
    line = dict(width = 2, color = "black")
    )))

fig.update_layout(
    title = "Idade x scores",
    xaxis_title = "Idades",
    yaxis_title = "Scores",
    showlegend = False
)

fig.show()
```

```python
# Error Bars

fig = go.Figure(data = go.Scatter(x = idades, y = scores, mode = "markers", marker = dict(
    color = "hotpink"
)))

fig.update_traces(error_y=dict(
    type = "data",
    array = [3,2,4,1,2,3],
    visible = True
))

fig.update_layout(
    title = "Erros Idades x Scores",
    xaxis_title = "Idades",
    yaxis_title = "Scores",
    showlegend = False
)

fig.show()
```

```python
cidades = ["Cidade A", "Cidade B", "Cidade C"]
populacao = [500000, 800000, 100000]
temperatura = [25, 30, 28]
PIB = [100000, 200000, 150000]
```

```python
# Bubble Chart

fig = go.Figure(data = go.Scatter(x = temperatura, y = populacao, mode = "markers", marker = dict(
    size = PIB,
    sizemode = "area",
    sizeref = max(PIB) / 10000,
    sizemin = 2,
    color = "hotpink",
    showscale = True
    ),
    text = cidades
    ))

fig.update_layout(
    title = "Relação Temperatura x PIB x População das Cidades",
    xaxis_title = "Temperatura",
    yaxis_title = "População",
    showlegend = False
)
fig.show()
```

```python
# Barras verticais

produtos = ["Produto A", "Produto B", "Produto C", "Produto D"]
vendas = [100, 200, 150, 300]
fig = go.Figure(data = go.Bar(x = produtos, y = vendas))

fig.update_layout(
    title = "Produtos x Vendas (vertical)",
    xaxis_title = "Produtos",
    yaxis_title = "Vendas"
)

fig.show()
```

```python
# Barras horizontais

fig = go.Figure(data=go.Bar(y = produtos, x = vendas, orientation="h"))
fig.update_layout(
    title = "Produtos x Vendas (horizontal)",
    xaxis_title = "Vendas",
    yaxis_title = "Produtos"
)
fig.show()
```

```python
# Stacked Bar Chart

meses = ["Janeiro", "Fevereiro", "Março"]
produtos = ["Produto A", "Produto B", "Produto C"]
vendas = [
    [100, 150, 200],
    [200, 250, 150],
    [150, 100, 200]
]
fig = go.Figure()
for i, produto in enumerate(produtos):
  fig.add_trace(go.Bar(
      x = meses,
      y = vendas[i],
      name = produto
  ))

  fig.update_layout(
      title = "Barras empilhadas Meses x Vendas dos produtos",
      xaxis_title = "Meses",
      yaxis_title = "Vendas"
  )

  fig.show()
```

```python
import numpy as np
from scipy.stats import norm

scores = [85, 90, 70, 92, 86, 88, 76, 95, 78, 80, 92, 86, 92, 88, 92]
fig = go.Figure(data=go.Histogram (x = scores, histnorm='probability density'))

x_range = np.linspace(min(scores), max(scores), 100)
y_gauss = norm.pdf(x_range, np.mean(scores), np.std(scores))

fig.add_trace(go.Scatter(
    x=x_range,
    y=y_gauss,
    mode='lines',
    name='Gaussiana',
    line=dict(color='red', width=3)
))

fig.update_layout(
    title = "Histograma Scores x Frequência",
    xaxis_title = "Scores",
    yaxis_title = "Frequência"
)
fig.show()
```

```python
# Pie Chart

categorias = ["Categoria A", "Categoria B", "Categoria C", "Categoria D"]
porcentagem_vendas = [30, 25, 20, 25]

colors = ["rgb(0, 132, 255)", "rgb(240, 29, 152)", "rgb(82, 255, 99)", "rgb(229, 255, 82)"]

labels = [f"{categoria}: {porcentagem}%"
for categoria, porcentagem in zip(categorias, porcentagem_vendas)]

fig = go.Figure(data = go.Pie(
    labels = labels,
    values = porcentagem_vendas,
    marker = dict(colors = colors)
))
fig.show()
```

```python
dados_vendas = {
    "Categoria A": {
        "Subcategoria 1": 40,
        "Subcategoria 2": 25,
        "Subcategoria 3": 35
    },

    "Categoria B": {
        "Subcategoria 1": 15,
        "Subcategoria 2": 40,
        "Subcategoria 3": 45
    },

    "Categoria C": {
        "Subcategoria 1": 30,
        "Subcategoria 2": 30,
        "Subcategoria 3": 40
    }
}
```

```python
# Sunburst Chart

fig = go.Figure(data = go.Sunburst(
    labels = list(dados_vendas.keys()) + list(dados_vendas['Categoria A'].keys()),
    parents=[''] * len(dados_vendas) + ['Categoria A'] * len(dados_vendas['Categoria A']),
    values = [sum(dados_vendas[categoria].values()) for categoria in dados_vendas] + list(dados_vendas['Categoria A'].values()),
))
fig.show()
```

```python
# Box Plot

categorias = ["Categoria A", "Categoria B", "Categoria C"]
dados = [[1, 2, 3, 4, 5], [2, 4, 6, 8, 10], [3, 6, 9, 12, 15]]
fig = go.Figure()
for categoria, dataset in zip(categorias, dados):
  fig.add_trace(go.Box(
      y = dataset,
      name = categoria
  ))
  fig.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
z = [3, 1, 4, 2, 5]
```

```python
# Visualização 3D

fig = go.Figure(data = go.Scatter3d(
    x = x,
    y = y,
    z = z,
    mode = "markers",
    marker = dict(
        size = 10,
        color = z,
        colorscale = "Viridis"
    )
))

fig.update_layout(
    scene = dict(
      xaxis = dict(
        title = "X-axis",
        range = [0, 6],
        tickmode = "linear",
        tick0 = 0,
        dtick = 1
    ),

    yaxis = dict(
        title = "Y-axis",
        range = [0, 6],
        tickmode = "linear",
        tick0 = 0,
        dtick = 1
    ),

    zaxis = dict(
        title = "Z-axis",
        range = [0, 6],
        tickmode = "linear",
        tick0 = 0,
        dtick = 1
    )
))

fig.show()
```
